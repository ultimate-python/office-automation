# การทำงานบน Excel

เป็นงานที่หลายคนใช้เวลาส่วนใหญ่ในการทำงานเพื่อจัดการให้เสร็จซ้ำๆ ไม่ว่าจะเป็นการวิเคราะห์ข้อมูล การรวมไฟล์ การคำนวนค่าต่างๆ รวมไปถึงการทำเป็นรายงาน ซึ่งข้อมูลของ Excel จะมีการเก็บแบบอ้างอิงด้วยคอมลัมภ์ และแถว ในการทำงานด้วย Python ต้องมีการตั้งชื่อคอลัมภ์ และแถวเพื่อให้เครื่องมือใน Python จัดการได้อย่างมีประสิทธิภาพ

เทคนิคการทำงานกับ Excel ด้วย Python คือ การจัดการชื่อคอลัมภ์ และชื่อแถวที่เก็บข้อมูลเดียวกันให้เหมือนกันทุกตัวอักษรรวมไปถึงเว้นวรรค เพื่อให้ Python จับคู่ข้อมูลที่ต้องทำงานร่วมกันได้อย่างถูกต้อง เพราะในการทำงานบน Python โดยเฉพาะเครื่องมือ pandas ที่เราใช้กันในตัวอย่างจะจับคู่ข้อมูลชื่อคอลัมภ์ และแถวแบบอัตโนมัติ ทำให้ถึงแม้ข้อมูลของเราอยู่สลับที่กัน ก็สามารถทำงานด้วยกันโดยที่ไม่อ้างอิงด้วยตำแหน่ง

## นำเข้าเครื่องมือที่ต้องการ

```python
import pandas as pd
import openpyxl
import os
```

## การดึงข้อมูล และสร้างตัวแปร

```python
file_folder = os.getcwd() + '/files'
all_file = os.listdir(file_folder)
report = pd.DataFrame()
```

```python
for file in all_file:
    table = pd.read_excel(file_folder + '/' + file,engine='openpyxl')
    table = table.set_index('Product')
    report = report.add(table,fill_value=0)
```

```python
report['All Year'] = report.sum(axis=1)
report.loc['All Product'] = report.sum(axis=0)
report.to_excel('Total Report.xlsx')
```
