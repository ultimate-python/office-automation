```python
import pandas as pd
import openpyxl
import os
```

##  

```python
file_folder = os.getcwd() + '/files'
all_file = os.listdir(file_folder)
report = pd.DataFrame()
```

```python
for file in all_file:
    table = pd.read_excel(file_folder + '/' + file,engine='openpyxl')
    table = table.set_index('Product')
    report = report.add(table,fill_value=0)
```

```python
report['All Year'] = report.sum(axis=1)
report.loc['All Product'] = report.sum(axis=0)
report.to_excel('Total Report.xlsx')
```
